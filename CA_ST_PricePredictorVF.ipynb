### 3. 4 Detecting Outliers <a name="3.4"></a>

According to Glen (2021), an outlier is a single or a set of datapoints that are considerably different from the other data points in the dataset. Sometimes they can be huge or very small.
It is possible to take a look in the outliers by using the Interquartile Range (IQR) technique, which measures where the middle fifty percent of the data are, or in other words, where the most of the middle values are (50%). For that, the method subtracts the third quartile (75%) from the first quartile(25%):
 
IQR = Q3 - Q1 
 
The outliers can be visualized through a box plot, which shows a box that are from the Q1 to Q3 quartile values of the data, and the line on the middle of the box is the median (Q2). The lines or as they are called, the whiskers, that are located at the edges of the box are the range of the data. And the separate dots are the outliers presented on the dataset.

```python
ford.plot(kind="box", subplots=True, figsize=(15,20), layout=(10,4))
plt.show()
```

With the function above we can take a look in all numerical variables and if they have or not outliers. And in the function below we can check the outliers of a specific column. 

```python
sns.boxplot(ford['engineSize'])
plt.show()
```

```python
Q1=ford.quantile(0.25)
Q3=ford.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
ford = ford[~((ford < (Q1-1.5 * IQR)) |(ford > (Q3 + 1.5 * IQR))).any(axis=1)]
```

### 7.5.4 Choosing the Optimal Number of Features<a name="7.5.4"></a>

By applying the Optimal Number of Features we can visualize how the r-square from the training and testing split will behave when we determine a range for the number of features. Here we chose a range that goes from 2 features until 20 features. (guys, we can change this range if we want, I just put an option).

```python
n_features_list = list(range(2,20))
adjusted_r2=[]
r2=[]
test_r2=[]

for n_features in range(2,20):
    LR = LinearRegression()
    
    rfe_n= RFE(LR, n_features_to_select=n_features)
    
    rfe_n.fit(X_train, y_train)
    
    col_n = X_train.columns[rfe_n.support_]
    
    X_train_rfe_n = X_train[col_n]
    
    X_train_rfe_n = sm.add_constant(X_train_rfe_n)
    
    LR_n = sm.OLS(y_train, X_train_rfe_n).fit()
    adjusted_r2.append(LR_n.rsquared_adj)
    r2.append(LR_n.rsquared)
    
    X_test_rfe_n = X_test[col_n]
    
    X_test_rfe_n = sm.add_constant(X_test_rfe_n, has_constant='add')
    
    y_pred = LR_n.predict(X_test_rfe_n)
    
    test_r2.append(r2_score(y_test, y_pred))
```

```python
plt.figure(figsize=(10,8))
plt.plot(n_features_list, adjusted_r2, label = 'adjusted_r2')
plt.plot(n_features_list, r2, label='train_r2')
plt.plot(n_features_list, test_r2, label='test_r2')
plt.legend(loc='upper left')
plt.show()
```

From the plot above, we can choose the number of features by taking in consideration the r2_score. 

We can see that as the number of features increases, so does the accuracy of r-square. For example, when we have only 4 features we get around 71% r-square accucary. On the other hand when we get 14, we get 83% for the training and 82.5% for the test.
